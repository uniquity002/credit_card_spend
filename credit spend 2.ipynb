```python
df_feature = data[['gender',
 'age',
 'lninc',
 'lncreddebt', 
    'ownvcr', 
    'owndvd',
   'owncd',

 'carvalue',
 'commutetime', 
 'reason',
 'card',
 'cardtype',
 'card2',
 'card2type',
 'total_spent']]
```

```python
from sklearn.model_selection import train_test_split

feature_train , feature_test ,label_train , label_test = train_test_split(df_feature.drop(columns='total_spent') , 
                                                                           df_feature['total_spent'] ,
                                                                           test_size = 0.3,
                                                                           random_state = 42)

```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

tuned_param = [{'max_features' : [2, 3,4 ] ,
               'n_estimators' : [100, 500,900]}]
random_forest_CV = GridSearchCV(RandomForestRegressor(),
                      tuned_param ,
                      cv=5 ,
                      n_jobs = -1 ,
                      verbose = 2)

random_forest_CV.fit(feature_train , label_train)

random_forest_CV.best_params_
```

```python


model =RandomForestRegressor(max_features=2 ,n_estimators=500)
model.fit(feature_train , label_train)
predict = model.predict(feature_train)

from sklearn.metrics import r2_score

print('train : ' , r2_score(label_train , predict))
predict_test = model.predict(feature_test)
print('test : ' , r2_score(label_test , predict_test))
from sklearn import metrics

print('rmse : trian = ' , np.sqrt(metrics.mean_squared_error(label_train , predict)))
print('rmse: test = ' , np.sqrt(metrics.mean_squared_error(label_test , predict_test)))
```
